```python
import requests

# Tampa Bay buoy - St. Petersburg station
url = "https://api.tidesandcurrents.noaa.gov/api/prod/datagetter"

params = {
    "station": "8726520",
    "product": "water_temperature",
    "begin_date": "20240101",
    "end_date": "20240131",
    "units": "metric",
    "time_zone": "lst",
    "format": "json",
    "interval": "h"
}

response = requests.get(url, params=params)
data = response.json()
print(data.keys())
```

```python
# Convert to a dataframe
df = pd.DataFrame(data['data'])
df['t'] = pd.to_datetime(df['t'])
df['v'] = pd.to_numeric(df['v'])

# Plot it
plt.figure(figsize=(12, 4))
plt.plot(df['t'], df['v'], color='steelblue', linewidth=1)
plt.title('Tampa Bay Water Temperature — January 2024 (St. Petersburg Buoy)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Data points: {len(df)}")
print(f"Temp range: {df['v'].min():.1f}°C to {df['v'].max():.1f}°C")
```

```python

# Pull data available at Tampa Bay stations
products = {
    "water_temperature": "Water Temp (°C)",
    "air_temperature": "Air Temp (°C)",
    "air_pressure": "Barometric Pressure (mb)"
}
dfs = {}

for product, label in products.items():
    resp = requests.get(url, params={**params, "product": product})
    result = resp.json()
    if 'data' in result:
        temp_df = pd.DataFrame(result['data'])
        temp_df['t'] = pd.to_datetime(temp_df['t'])
        temp_df['v'] = pd.to_numeric(temp_df['v'], errors='coerce')
        dfs[label] = temp_df
        print(f"✓ {label}: {len(temp_df)} points")
    else:
        print(f"✗ {label}: not available — {result.get('error', {}).get('message', 'unknown error')}")

        # Wind needs special handling - different column structure
wind_resp = requests.get(url, params={**params, "product": "wind"})
wind_raw = pd.DataFrame(wind_resp.json()['data'])
wind_raw['t'] = pd.to_datetime(wind_raw['t'])
wind_raw['s'] = pd.to_numeric(wind_raw['s'], errors='coerce')  # speed
wind_raw['d'] = pd.to_numeric(wind_raw['d'], errors='coerce')  # direction degrees
wind_raw['g'] = pd.to_numeric(wind_raw['g'], errors='coerce')  # gusts

# Add to our dfs dictionary with consistent naming
dfs["Wind Speed (m/s)"] = wind_raw.rename(columns={'s': 'v'})[['t', 'v']]
dfs["Wind Direction"] = wind_raw.rename(columns={'d': 'v'})[['t', 'v']]
dfs["Wind Gusts (m/s)"] = wind_raw.rename(columns={'g': 'v'})[['t', 'v']]

print(f"✓ Wind data: {len(wind_raw)} points")
print(f"Speed range: {wind_raw['s'].min():.1f} to {wind_raw['s'].max():.1f} m/s")
print(f"Gust max: {wind_raw['g'].max():.1f} m/s")
print(f"\nAll available data streams: {list(dfs.keys())}")
```

```python
print("Data streams loaded:")
for name, df in dfs.items():
    print(f"  ✓ {name}: {len(df)} points")
```

```python
fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

plots = [
    ("Water Temp (°C)", "steelblue"),
    ("Air Temp (°C)", "tomato"),
    ("Barometric Pressure (mb)", "purple"),
    ("Wind Speed (m/s)", "darkorange"),
]

for ax, (name, color) in zip(axes, plots):
    ax.plot(dfs[name]['t'], dfs[name]['v'], color=color, linewidth=0.8)
    ax.set_ylabel(name, fontsize=9)
    ax.grid(True, alpha=0.3)

axes[0].set_title('Tampa Bay Environmental Conditions — January 2024 (St. Petersburg)', 
                   fontsize=12, fontweight='bold')
axes[-1].set_xlabel('Date')
plt.tight_layout()
plt.show()
```
